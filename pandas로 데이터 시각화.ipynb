```python
import pandas as pd
```

```python
df_kosis = pd.read_csv('data\population.csv', encoding='cp949')
df_kosis.shape
```

```python
pd.options.display.max_columns = 793
```

```python
df_kosis.head()
```

```python
df_kosis.head()
```

```python
df = df_kosis.melt(id_vars='시군구별')
df.head()
```

```python
df['시군구별'].unique()
```

```python
print(df.shape)
df = df[df['시군구별'] != '시군구별'].copy()
df.shape
```

```python
df.head()
```

```python
df['연도'] = df['variable'].str.split('.', expand=True)[0]
df['월'] = df['variable'].str.split('.', expand=True)[1]
df['성별'] = df['variable'].str.split('.', expand=True)[2]
df.head()
```

```python
df.tail()
```

```python
df['성별'].unique()
```

```python
df['성별'].nunique()
```

```python
df['성별'] = df['성별'].fillna('전체')
df['성별'].unique()
```

```python
df['성별'] = df['성별'].replace('1', '남자').replace('2', '여자')
```

```python
df['성별'].unique()
```

```python
df['성별'].value_counts()
```

```python
df.head()
```

```python
df = df.rename(columns={'variable':'기간', 'value':'출생아수'})
df.head()
```

```python
df_all = df[(df['시군구별']=='전국') & (df['성별']=='전체')]
df_all.head(2)
```

```python
df_all = df_all[['연도', '월', '출생아수']].copy()
df_all.head()
```

```python
import matplotlib.pyplot as plt
plt.rc("font", family='Malgun Gothic')
```

```python
df_all['출생아수'] = pd.to_numeric(df_all['출생아수'])
```

```python
df_all.set_index(['연도', '월']).plot(figsize=(15, 4))
```

```python
df_all[-24:].set_index(['연도', '월']).plot.bar(figsize=(15, 4))
```
